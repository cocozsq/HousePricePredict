```python
import pandas as pd
```

```python
# save filepath to variable for easier access
melbourne_file_path = 'data/melb_data/melb_data.csv'

#read the data and store data in DataFrame titled melbourne_data
melbourne_data = pd.read_csv(melbourne_file_path)

# print a summary of the data in Melbourne data
melbourne_data.describe()
```

```python
melbourne_data = melbourne_data.dropna(axis=0)
```

```python
melbourne_data.head(5)
```

```python
# Selecting The Prediction Target
y = melbourne_data.Price

# Choosing "Features"
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']

X = melbourne_data[melbourne_features]

X.describe()
```

```python
X.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
# Define model. Specify a number for random_state to ensure some results each run
melbourne_model = DecisionTreeRegressor(random_state = 1)

melbourne_model.fit(X, y)
```

```python
print("Making predictions for the following 5 houses")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
```

```python
# 计算MAE 平均绝对差
from sklearn.metrics import mean_absolute_error
predicted_home_prices = melbourne_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

## Validation

```python
from sklearn.model_selection import train_test_split
# split data into training and validation , for both features and target
# The split is based on a random number generator. Supplying a numeric value to the
# argument gyarantees we get the same split every time we run this script
X_train, x_test, y_train, y_test = train_test_split(X, y, random_state=0)

melbourne_model = DecisionTreeRegressor()
melbourne_model.fit(X_train, y_train)
# get predicted home price
predict_y = melbourne_model.predict(x_test)

from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test, predict_y))
```

### overfitting and underfitting

#### Models can suffer from either:
###### Overfitting: capturing spurious patterns that won't recur in the future, leading to less predictions,
###### Underfitting: failing to capture relevant patterns, again leading to less accuracy predictions

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, train_X,val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes = max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(preds_val, val_y)
    return mae

for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, X_train, x_test, y_train, y_test)
    print("   Max leaf nodes:%d \t\t Mean Absolute Error: %d" %(max_leaf_nodes,my_mae))
    
```
